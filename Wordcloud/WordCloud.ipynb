```
text_file = open("wordcloud.txt")
text = text_file.read()
```

```
import nltk
nltk.download('all')
from nltk import sent_tokenize
from nltk import word_tokenize
```

```
sentences = sent_tokenize(text)
words = word_tokenize(text)
from nltk.probability import FreqDist
fdist = FreqDist(words)
fdist.most_common(10)
```

```
words_no_punc = []
for w in words:
  if w.isalpha():
    words_no_punc.append(w.lower())
```

```
from nltk.corpus import stopwords
stopwords = stopwords.words("english")
list = ['\'',',']
stopwords=stopwords+list
print(stopwords)
```

```
fdist = FreqDist(words_no_punc)
from nltk.corpus import stopwords
stopwords = stopwords.words("english")
clean_words=[]
for w in words_no_punc:
  if w not in stopwords:
    clean_words.append(w)
fdist = FreqDist(clean_words)
```

```
from nltk.stem import PorterStemmer
stemma = []
porter=PorterStemmer()
word_list=clean_words
for w in word_list:
  stemma.append(print(porter.stem(w)))
str1=' '.join([str(elem) for elem in clean_words])
```

```
from wordcloud import WordCloud
import matplotlib.pyplot as plt
wordcloud = WordCloud(mode="RGB").generate(str1)
plt.figure(figsize = (12,12))
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
wordcloud.to_file("jan1.png")
#plt.savefig('jan.png')
#files.download('jan.png')
```
